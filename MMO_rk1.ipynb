```python
from sklearn.preprocessing import RobustScaler
```

```python
data.head()
```

```python
data['Price_euros'].plot.hist(title='Price histogram');
```

```python
scaler = RobustScaler()
scaled_data = pd.Series(scaler.fit_transform(data[['Price_euros']]).reshape(1, -1)[0])
scaled_data.plot.hist(title='Scaled price histogram');
```

```python
data['Price_euros'].median()
```

```python
scaled_data.median()
```
